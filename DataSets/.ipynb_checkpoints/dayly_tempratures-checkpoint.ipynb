```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('daily_tempratures.csv')
df = df.drop(['Unnamed: 2'],axis=1)
df.columns = ['date', 'value']
df.drop(df.tail(2).index,inplace=True) # drop last n rows
df['date'] = pd.to_datetime(df['date'])
#vì kiểu dữ liệu đều là object nên col value -> float,date -> datetime64
df['value'] = pd.to_numeric(df.value, errors='coerce')
df.set_index('date', inplace=True)
print('\nTypefo date:')
print(df.dtypes)
plt.figure(figsize=(25,5))
plt.plot(df)
plt.show()
```

```python
# nhận thấy dữ liệu chỉ có dạng Seadsonal, và mỗi Seasonal có khoảng tg là 1 năm
# sử dụng differencing để loại bỏ seasonal 
diff = list()
days_in_year = 365
for i in range(days_in_year, len(df.value)):
    value = df.value[i] - df.value[i - days_in_year]
    diff.append(value) 
plt.figure(figsize=(15,5))
plt.plot(diff)
plt.show()

#######################################
# ts_log = np.log(df)
# plt.figure(figsize=(15,5))
# plt.plot(ts_log)
# plt.show()
```
